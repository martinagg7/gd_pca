```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Lista de variables continuas
continuous_cols = [
    'Edad', 'Experiencia_laboral', 'Gastoscore', 'Family_Size', 'Generacion',
    'Ingresos_anuales_brutos', 'Digital_encuesta',
    'Gasto_medio_mensual_cuota', 'Dias_cliente'
]

# Asegúrate de que estas columnas estén en tipo numérico
for col in continuous_cols:
    df_pca[col] = pd.to_numeric(df_pca[col], errors='coerce')

# Diccionario para almacenar cuántos registros se eliminan por columna
eliminados_por_columna = {}

# Función para eliminar outliers por IQR
def eliminar_outliers_iqr(df, columnas):
    df_filtrado = df.copy()
    for col in columnas:
        Q1 = df_filtrado[col].quantile(0.1)
        Q3 = df_filtrado[col].quantile(0.9)
        IQR = Q3 - Q1
        limite_inferior = Q1 - 1.5 * IQR
        limite_superior = Q3 + 1.5 * IQR

        # Contar cuántos valores están fuera de los límites en el dataset original
        fuera_de_limites = ~((df[col] >= limite_inferior) & (df[col] <= limite_superior))
        eliminados_por_columna[col] = fuera_de_limites.sum()

        # Filtrar en el DataFrame
        df_filtrado = df_filtrado[
            (df_filtrado[col] >= limite_inferior) & (df_filtrado[col] <= limite_superior)
        ]
    return df_filtrado

# Aplicar la función
df_sin_outliers = eliminar_outliers_iqr(df_pca, continuous_cols)

# Mostrar resultado general
print(f"Antes: {df_pca.shape}")
print(f"Después de eliminar outliers: {df_sin_outliers.shape}\n")

# Mostrar registros eliminados por columna
print("Registros eliminados por columna:")
for col, count in eliminados_por_columna.items():
    print(f"- {col}: {count}")
```

# PCA


El **PCA** permite reducir la dimensionalidad del conjunto de datos manteniendo la mayor cantidad de varianza posible. En esta sección se busca entender cómo se distribuye la información entre los componentes principales y qué variables aportan más a cada uno.

1. Se aplica PCA utilizando todas las variables continuas.
2. El objetivo es maximizar la varianza explicada con el menor número de componentes.
3.  Se analizan las cargas de los componentes para estudiar la relevancia de cada variable.

**PCA Continuas**

```python
print("Columnas que se usarán en el PCA:")
print(continuous_cols)
```

```python

scaler = StandardScaler()
X_cont_scaled = scaler.fit_transform(df_sin_outliers[continuous_cols])

# Crear DataFrame estandarizado
df_pca_input_cont = pd.DataFrame(X_cont_scaled, columns=continuous_cols, index=df_sin_outliers.index)


print(" Variables incluidas en PCA (solo continuas):")
print(df_pca_input_cont.columns.tolist())

# PCA
pca = PCA(n_components=2)
X_pca_cont = pca.fit_transform(df_pca_input_cont)

# Resultados
df_pca_result_cont = pd.DataFrame(X_pca_cont, columns=['PC1', 'PC2'], index=df_sin_outliers.index)

# Varianza explicada
explained_var = pca.explained_variance_ratio_ * 100
print(f"\n Varianza explicada por PC1: {explained_var[0]:.2f}%")
print(f"Varianza explicada por PC2: {explained_var[1]:.2f}%")
print(f" Varianza total acumulada: {explained_var.sum():.2f}%")

# Gráfico
plt.figure(figsize=(8, 6))
plt.scatter(df_pca_result_cont['PC1'], df_pca_result_cont['PC2'], alpha=0.5)
plt.xlabel(f'PC1 ({explained_var[0]:.2f}%)')
plt.ylabel(f'PC2 ({explained_var[1]:.2f}%)')
plt.title('PCA con solo variables continuas')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Cargas Variables

```python
loadings = pd.DataFrame(
    pca.components_.T,
    columns=['PC1', 'PC2'],
    index=df_pca_input_cont.columns  
)


print("Cargas de cada variable en los componentes:")
loadings
```

```python
plt.figure(figsize=(10, 6))
loadings['PC1'].sort_values().plot(kind='barh', color='skyblue')
plt.title(f'Cargas de las variables en PC1 ({explained_var[0]:.2f}%)')
plt.xlabel('Carga')
plt.grid(True)
plt.tight_layout()
plt.show()

# Gráfico de cargas - PC2
plt.figure(figsize=(10, 6))
loadings['PC2'].sort_values().plot(kind='barh', color='salmon')
plt.title(f'Cargas de las variables en PC2 ({explained_var[1]:.2f}%)')
plt.xlabel('Carga')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# K-Means

Una vez proyectados los datos en el espacio reducido  (PCA), procedemos a segmentar a los clientes de la empresa con el objetivo de identificar **patrones de comportamiento comunes**.

Para determinar el número óptimo de clusters, aplicamos tres enfoques complementarios: 

1) Método de la inercia , donde evaluamos la suma de las distancias  los puntos a sus respectivos centroides
2) Método del coeficiente de silueta, que mide la cohesión y separación entre clusters 
3) Una revisión manual de la proporción de clientes en cada cluster, con el fin de asegurarnos de que los grupos estén bien distribuidos y sean estratégicamente útiles.

## Método del Codo

El método del codo se utiliza para determinar el número óptimo de clusters $ k $ en un modelo de clustering como K-means. Se basa en observar la evolución de la **inercia intra-cluster**, es decir, la suma de las distancias cuadradas entre los puntos y el centroide de su clúster.

La fórmula de la inercia total es:

$
\text{Inercia} = \sum_{i=1}^{k} \sum_{x \in C_i} \|x - \mu_i\|^2
$

donde:
- $ k $: número de clusters
- $ C_i $: conjunto de puntos en el cluster $ i $
- $mu_i $: centroide del cluster $ i $
- $|x - \mu_i\|^2$: distancia euclídea al cuadrado

A medida que aumenta $ k $, la inercia disminuye, pero llega un punto donde la reducción se vuelve marginal. Ese punto de inflexión —el "codo" de la curva— indica un buen valor de $k$ que equilibra simplicidad y calidad del modelo.

```python
inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_pca_result_cont[['PC1', 'PC2']])
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, 'o-', color='darkblue')  
plt.xlabel('Número de clusters')
plt.ylabel('Inercia')
plt.title('Método del Codo')
plt.grid(True)
plt.show()
```

Conclusión: Según el método del codo, **el número óptimo de clusters es 5**

## División final Clusters

```python
# Aplicar K-Means
kmeans = KMeans(n_clusters=5, random_state=42)
df_pca_result_cont['cluster'] = kmeans.fit_predict(df_pca_result_cont[['PC1', 'PC2']])

# Centroides
centroids = kmeans.cluster_centers_

# Gráfico
plt.figure(figsize=(10, 8))
sns.scatterplot(
    data=df_pca_result_cont,
    x='PC1', y='PC2',
    hue='cluster',
    palette='Set2',
    alpha=0.7,
    s=60
)

# Añadir los centroides
plt.scatter(
    centroids[:, 0], centroids[:, 1],
    c='black', s=20, marker='x', label='Centroides'
)

plt.title('Clientes segmentados por K-Means (k=5) en el espacio PCA')
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.grid(True)
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()
```

## Método de la Silueta

```python
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm
import numpy as np

# Datos y etiquetas
X = df_pca_result_cont[['PC1', 'PC2']].values
labels = df_pca_result_cont['cluster']
n_clusters = len(np.unique(labels))

# Calcular silueta individual
silhouette_vals = silhouette_samples(X, labels)

# Gráfico
plt.figure(figsize=(17, 6))
y_lower = 10

for i in range(n_clusters):
    ith_cluster_silhouette_vals = silhouette_vals[labels == i]
    ith_cluster_silhouette_vals.sort()
    
    size_cluster_i = ith_cluster_silhouette_vals.shape[0]
    y_upper = y_lower + size_cluster_i
    
    color = cm.Set2(i / n_clusters)
    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_vals,
                      facecolor=color, edgecolor=color, alpha=0.8)
    
    plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10  # espacio entre clusters

# Promedio global
plt.axvline(x=silhouette_score(X, labels), color="red", linestyle="--", label="Media global")

# Estética
plt.title("Método del Silueta")
plt.xlabel("Coeficiente de silueta")
plt.ylabel("Observaciones ordenadas")
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

 - Todos los clusters tienen formas anchas y positivas, lo cual indica buena **cohesión interna**.
- Hay muy pocas observaciones con silueta negativa, concentradas en el **cluster 0**, lo que sugiere la posible presencia de puntos en el borde o mal asignados, pero no es preocupante.
- **Clusters 4 y 3** tienen las mejores formas: compactos, bien definidos y lejanos entre sí.

**Conclsuión**: El modelo de **k = 5** ofrece una segmentación **sólida y balanceada**, con grupos bien diferenciados en su mayoría.
